```python
def prepare_input(claims: str, description: str, abstract: str) -> str:
    # Concatenate claims, description, and abstract as a single input string
    input_text = f"Claims: {claims} Description: {description} Abstract: {abstract}"
    return input_text
```

```python
def tokenize_input(text: str):
    # Tokenize the text
    inputs = tokenizer(text, padding=True, truncation=True, return_tensors="pt")
    return inputs
```
